# MatPlotLib

### It is a visualisation Library in python. Provides API to embedding plots into applications

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
## creating arrays

x = np.arange(0,10)
y = np.arange(11,21)
```

```python
## plotting using matplotlib

## scatter - scatter values of x and y

plt.scatter(x,y,c="r")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph")
```

```python
y=x*x
```

```python
## plot - used to plot straight line for given values

plt.plot(x,y,"r",linestyle="dashed", linewidth=2, markersize=12)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Plotted graph")
```

```python
## Subplot - mutiple plots within same plot

plt.subplot(2,2,1)
plt.plot(x,y,"r--")
plt.subplot(2,2,2)
plt.plot(x,y,"b")
plt.subplot(2,2,3)
plt.plot(x,y,"g")
```

```python
## Plotting a sin curve

x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title("Sin curve")

plt.plot(x,y)
```

```python
## Sin and cos curves

x = np.arange(0, 4*np.pi,0.1)
y_cos = np.cos(x)
y_sin = np.sin(x)

## Subplot

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('Sin curve')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cos curve')
```

```python
## Bar plot

x=[1,4,8]
y=[2,5,9]

x2=[3,6,7]
y2=[11,14,17]

plt.bar(x,y)
plt.bar(x2,y2,color="g")
```

```python
## Histograms

a = np.array(np.random.randint(0,50,15))

plt.hist(a)
```

## Box Plot 

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert=True, patch_artist=True )
```
