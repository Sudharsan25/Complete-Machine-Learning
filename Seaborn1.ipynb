# Seaborn

### * distplot
### * joinplot
### * pairplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.corr()
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())
```

```python
## Joinplot()

sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.pairplot(df, hue='sex')
```

```python
sns.displot(df['tip'])
```

```python
## If kde is set to true then the Y-axis shows density instead of count

sns.displot(df['tip'], kde=True, bins=50)
```

## Categorical plots

```python
sns.countplot(x='day',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(data=df,orient='h')
```

```python
sns.boxplot(x='sex',y='tip',data=df)
```

```python
sns.violinplot(x='day',y='total_bill',data=df, palette='rainbow')
```
